Regresin lineal polinmica
==========================

Es un caso especial de la regresin lineal simple con el cual adquirimos flexibilidad para encontrar relaciones no lineales. El modelo est dado por:

<center>

$\hat{Y_i} = \hat{\beta_0} + \hat{\beta_1}X_i + \hat{\beta_1}X_i + ... + \hat{\beta_1}X_i^n$



</center>

Se dice que este modelo es un caso especial puesto que a la variable $X_i$ se le aplica una transformacin que genera "ms variables" y con esto podemos seguir expresando a $Y$ con una combinacin lineal (de ah proviene el nombre).

Implementacin
==========================

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el dataset
dataset = pd.read_csv('./datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
Y = dataset.iloc[:, 2].values
```

```python
# Aplicamos la transformacin polinmica
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)

# Comparemos X y X_poly
print("X: \n")
print(X, "\n")
print("X_poly: \n")
print(X_poly, "\n")
```

```python
# Ajustamos el modelo de regresin lineal con los datos polinmicos
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_poly, Y)

# Graficamos los resultados obtenidos
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = "red")
plt.plot(X_grid, regressor.predict(poly_reg.fit_transform(X_grid)), color = "blue")
plt.title("Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```
