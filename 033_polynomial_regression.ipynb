¿Qué es la regresión polinomial?
==========================

Es un caso especial de la regresión lineal simple con el cual adquirimos flexibilidad para encontrar relaciones no lineales. El modelo está dado por:

<center>

$\hat{Y_i} = \hat{\beta_0} + \hat{\beta_1}X_i + \hat{\beta_1}X_i² + ... + \hat{\beta_1}X_i^n$

</center>

Se dice que este modelo es un caso especial puesto que a la variable $X_i$ se le aplica una transformación que genera "más variables" y con esto podemos seguir expresando a $Y$ con una combinación lineal (de ahí proviene el nombre).

Implementación
==========================
