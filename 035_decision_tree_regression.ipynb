rboles de decisin para regresin
==========================

Son un modelos predictivos formados por reglas binarias con las que se consige repartir las observaciones en funcin de sus atributos y as realizar predicciones. De manera recursiva se van haciendo divisiones con el conjunto de datos maximizando la cantidad de puntos similares por seccin y a la hora de realizar la prediccin retornamos la media. Implementacin
==========================

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importamos el dataset
dataset = pd.read_csv('./datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Entrenamos el modelo
from sklearn.tree import DecisionTreeRegressor
regression = DecisionTreeRegressor(random_state = 0)
regression.fit(X, y)
```

```python
# Visualizacin de los resultados
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = "red")
plt.plot(X_grid, regression.predict(X_grid), color = "blue")
plt.title("Decision Tree Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```
