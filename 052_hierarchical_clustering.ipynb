Clustering jerárquico
=========================

Es un algoritmo aglomerativo, el cual comienza con una cantidad de clusters igual al número de puntos, y conforme avanza en las iteraciones une aquellos que tengan la menor distancia entre sí hasta formar uno solo. Para seleccionar el número k de clusters, se construye el dendograma (un diagrama de árbol que muestra cómo los grupos fueron creándose y la distancia que había entre ellos) y se busca cortar la línea vertical que tenga la mayor distancia entre horizontales.

Implementación
=========================

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importamos el dataset
dataset = pd.read_csv('./datasets/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
# Creación del dendograma
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = "ward"))
plt.title("Dendrograma")
plt.show()
```

```python
# Aplicamos el clustering jerárquico
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
y_hc = hc.fit_predict(X)
```

```python
# Visualizamos los clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 50, c = "red")
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 50, c = "blue")
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 50, c = "green")
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 50, c = "cyan")
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 50, c = "magenta")
plt.legend()
plt.show()
```
