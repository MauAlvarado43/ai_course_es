Clasificacin
=========================

A diferencia de la regresin que busca predecir un valor cuantitativo, en los problemas de clasificacin el enfoque tiende en torno a una variable cualitativa. Cabe destacar que la salida de los modelos es una probabilidad de pertenencia a cada categora.

Rendimiento
=========================

Un concepto clave en la evaluacin de los modelos de clasificacin es la matriz de confusin, esta nos permite comparar los valores predichos contra los valores reales.

Ahora bien, con esta matriz podemos calcular coeficientes que nos ayudan a analizar los resultados generados:

* Exactitud (Accuracy): hace referencia a qu tan cerca est el resultado de la observacin. $\frac{VP + VN}{VP + FP + FN + VN}$

* Precisin (Precision): se refiere a la dispersin del conjunto de valores obtenidos. $\frac{VP}{VP + FP}$

* Sensibilidad (Recall): es la proporcin de casos positivos calculados correctamente. $\frac{VP}{VP + FN}$

* Especificidad (Especificity): son los catos negativos que el algoritmo clasific correctamente. $\frac{VN}{VN + FP}$

* F1 Score: dado que la precisin y el recall por si solos pueden ser engaosos al haber una amplia desigualdad en las clases, esta medida nos permite evaluar el rendimiento sin verse afectada por este problema. $\frac{2*Precision*Recall}{Precision + Recall}$
