Clasificación
=========================

A diferencia de la regresión que busca predecir un valor cuantitativo, en los problemas de clasificación el enfoque tiende en torno a una variable cualitativa. Cabe destacar que la salida de los modelos es una probabilidad de pertenencia a cada categoría.

Algoritmos
=========================
Para esta sección veremos los siguientes algoritmos:

* Regresión logística
* KNN
* Máquinas de soporte vectorial
* Naive Bayes
* Árboles de decisión
* Bosques aleatorios de regresión

Rendimiento
=========================

Un concepto clave en la evaluación de los modelos de clasificación es la matriz de confusión, esta nos permite comparar los valores predichos contra los valores reales.

Ahora bien, con esta matriz podemos calcular coeficientes que nos ayudan a analizar los resultados generados:

* Exactitud (Accuracy): hace referencia a qué tan cerca está el resultado de la observación. $\frac{VP + VN}{VP + FP + FN + VN}$

* Precisión (Precision): se refiere a la dispersión del conjunto de valores obtenidos. $\frac{VP}{VP + FP}$

* Sensibilidad (Recall): es la proporción de casos positivos calculados correctamente. $\frac{VP}{VP + FN}$

* Especificidad (Especificity): son los catos negativos que el algoritmo clasificó correctamente. $\frac{VN}{VN + FP}$

* F1 Score: dado que la precisión y el recall por si solos pueden ser engañosos al haber una amplia desigualdad en las clases, esta medida nos permite evaluar el rendimiento sin verse afectada por este problema. $\frac{2*Precision*Recall}{Precision + Recall}$
